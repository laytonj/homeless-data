## Format, Clean Up

*Remove extra rows created full of NaN.*

```python
for file in pitFilesCoc:
    file.dropna(how='all', inplace=True)
pit2017Coc.tail(n=2)
```

*Remove the Total row*

```python
for file in pitFilesCoc:
    file.reset_index(inplace=True) #set as numeric index
    file.set_index('CoC Name', inplace=True) #set index to CoC name
    file.drop('Total', inplace=True) #drop where has no CoC name (the totals row is the only one)
    file.reset_index(inplace=True) #reset index back to numeric
    file.set_index('CoC Number', inplace=True) #set index to CoC number
    
pit2017Coc.tail(n=3)
```

*Remove the , YYYY from each column header*

```python
i = 0
for file in pitFilesCoc:
    year = str(years[i])
    replaceThis = ', '+year
    file.columns = file.columns.map(lambda x: x.replace(replaceThis,''))
    i+=1
pit2017Coc.head(n=3)
```

*Remove commas, convert to numeric*

```python
#This is slightly different from the state PIT syntax because at least one row was given cell value of '.', 
#That value must be removed to NaN, so float will then need to be used rather than int.
def removeComma(x):
    if type(x) == str:
        x = x.replace(',', '')
        if x.isnumeric() == False:
            x = np.nan
    return float(x)

'''
Looping through seems to replace the object. "pit2007st is files[0]" would then return False, whereas True until now.
In other words, the change would have taken place for files[0] but not for pit2007st.
Same problem with a syntax like 'for file in files:'

for i in range(len(files)):
    files[i] = files[i].applymap(lambda x: removeComma(x))
    
Why???    
'''
```

```python
pit2007Coc.iloc[:,2:] = pit2007Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2008Coc.iloc[:,2:] = pit2008Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2009Coc.iloc[:,2:] = pit2009Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2010Coc.iloc[:,2:] = pit2010Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2011Coc.iloc[:,2:] = pit2011Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2012Coc.iloc[:,2:] = pit2012Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2013Coc.iloc[:,2:] = pit2013Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2014Coc.iloc[:,2:] = pit2014Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2015Coc.iloc[:,2:] = pit2015Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2016Coc.iloc[:,2:] = pit2016Coc.iloc[:,2:].applymap(lambda x: removeComma(x))
pit2017Coc.iloc[:,2:] = pit2017Coc.iloc[:,2:].applymap(lambda x: removeComma(x))

pit2017Coc.head(n=3)
```

*For earlier years when all chronically homeless were individuals, fill in total chronically homeless as individuals*

```python
#what looks like before...
pitAllCoc.loc[['CA-500'],['year','totalChronic','shelChronic','unshelChronic','totalChronicIndiv','shelChronicIndiv','unshelChronicIndiv']]
```

```python
pitAllCoc['totalChronic'] = np.where(pitAllCoc['totalChronic'].isnull(), pitAllCoc['totalChronicIndiv'], pitAllCoc['totalChronic'])
pitAllCoc['shelChronic'] = np.where(pitAllCoc['shelChronic'].isnull(), pitAllCoc['shelChronicIndiv'], pitAllCoc['shelChronic'])
pitAllCoc['unshelChronic'] = np.where(pitAllCoc['unshelChronic'].isnull(), pitAllCoc['unshelChronicIndiv'], pitAllCoc['unshelChronic'])

pitAllCoc.loc[['CA-500'],['year','totalChronic','shelChronic','unshelChronic','totalChronicIndiv','shelChronicIndiv','unshelChronicIndiv']]
```

*Create non-CH and non-Vets columns*

```python
pitAllCoc['totalNonChronic'] = pitAllCoc['totalH']-pitAllCoc['totalChronic']
pitAllCoc['shelNonChronic'] = pitAllCoc['shelH']-pitAllCoc['shelChronic']
pitAllCoc['unshelNonChronic'] = pitAllCoc['unshelH']-pitAllCoc['unshelChronic']

pitAllCoc.tail(n=3)
```

```python
pitAllCoc['totalNonVet'] = pitAllCoc['totalH']-pitAllCoc['totalVet']
pitAllCoc['shelNonVet'] = pitAllCoc['shelH']-pitAllCoc['shelVet']
pitAllCoc['unshelNonVet'] = pitAllCoc['unshelVet']-pitAllCoc['unshelVet']

pitAllCoc.tail(n=3)
```


*Because the columns had been automatically sorted alphabetically above, revert the order*

```python
pitAllCoc = pitAllCoc[['cocName', 'st', 'year', 'totalH', 'shelH', 'unshelH', 
                     'totalIndiv', 'shelIndiv', 'unshelIndiv', 
                     'totalInFam', 'shelInFam', 'unshelInFam',
                     'totalChronic', 'shelChronic', 'unshelChronic', 
                     'totalNonChronic', 'shelNonChronic', 'unshelNonChronic',
                     'totalChronicIndiv', 'shelChronicIndiv', 'unshelChronicIndiv', 
                     'totalChronicInFam', 'shelChronicInFam', 'unshelChronicInFam', 
                     'totalVet', 'shelVet', 'unshelVet', 
                     'totalNonVet','shelNonVet','unshelNonVet', 
                     'totalYouthAll', 'shelYouthAll', 'unshelYouthAll', 
                     'totalYouthChild', 'shelYouthChild', 'unshelYouthChild', 
                     'totalYouthYA', 'shelYouthYA', 'unshelYouthYA', 
                     'totalParYouthAll', 'shelParYouthAll', 'unshelParYouthAll', 
                     'totalParYouthChild', 'shelParYouthChild', 'unshelParYouthChild', 
                     'totalParYouthYA', 'shelParYouthYA', 'unshelParYouthYA', 
                     'totalChildOfParYouth', 'shelChildOfParYouth', 'unshelChildOfParYouth']]
pitAllCoc.head(n=3)
```
