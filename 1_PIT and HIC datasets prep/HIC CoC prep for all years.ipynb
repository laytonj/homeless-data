*Remove commas, convert to numeric*

```python
def removeComma(x):
    if type(x) == str:
        x = x.replace(',', '')
        if x.isnumeric() == False:
            x = np.nan
    return float(x)

'''As seen with the PIT prep syntax, using looping through the files to implement applymap doesn't seem to work. 
Not sure why.'''
```

# Create an all-years HIC DataFrame

*Combine all years*

```python
hicAllCoc = pd.concat(hicFilesCoc, axis=0, join="outer")
hicAllCoc.tail(3)
#note it sorted columns in alphabetical order
```

```python
hicAllCoc.columns.values
```

*Create columns grouping temporary beds and permanent beds*

```python
hicAllCoc['permBedsAll'] = hicAllCoc.fillna(0)['pshBedsAll'] + hicAllCoc.fillna(0)['rrhBedsAll']
hicAllCoc['permBedsWithChildren'] = hicAllCoc.fillna(0)['pshBedsWithChildren'] + hicAllCoc.fillna(0)['rrhBedsWithChildren']
hicAllCoc['permBedsNoChildren'] = hicAllCoc.fillna(0)['pshBedsNoChildren'] + hicAllCoc.fillna(0)['rrhBedsNoChildren']
hicAllCoc['permBedsOnlyChildren'] = hicAllCoc.fillna(0)['pshBedsOnlyChildren'] + hicAllCoc.fillna(0)['rrhBedsOnlyChildren']
hicAllCoc['permUnitsWithChildren'] = hicAllCoc.fillna(0)['pshUnitsWithChildren'] + hicAllCoc.fillna(0)['rrhUnitsWithChildren']

hicAllCoc['tempBedsAll'] = hicAllCoc.fillna(0)['esBedsAll'] + hicAllCoc.fillna(0)['thBedsAll'] + hicAllCoc.fillna(0)['shBedsAll']
hicAllCoc['tempBedsWithChildren'] = hicAllCoc.fillna(0)['esBedsWithChildren'] + hicAllCoc.fillna(0)['thBedsWithChildren'] + hicAllCoc.fillna(0)['shBedsWithChildren']
hicAllCoc['tempBedsNoChildren'] = hicAllCoc.fillna(0)['esBedsNoChildren'] + hicAllCoc.fillna(0)['thBedsNoChildren'] + hicAllCoc.fillna(0)['shBedsNoChildren']
hicAllCoc['tempBedsOnlyChildren'] = hicAllCoc.fillna(0)['esBedsOnlyChildren'] + hicAllCoc.fillna(0)['thBedsOnlyChildren'] + hicAllCoc.fillna(0)['shBedsOnlyChildren']
hicAllCoc['tempUnitsWithChildren'] = hicAllCoc.fillna(0)['esUnitsWithChildren'] + hicAllCoc.fillna(0)['thUnitsWithChildren'] + hicAllCoc.fillna(0)['shUnitsWithChildren']
```

*Create a column for sum of all beds*

```python
hicAllCoc['allBeds'] = hicAllCoc['permBedsAll'] + hicAllCoc['tempBedsAll']
```

*Revise the order of columns because they had been automatically sorted alphabetically above*

```python
hicAllCoc = hicAllCoc[['st', 'year', 'allBeds',
                      'tempBedsAll', 'permBedsAll', 'esBedsAll', 'thBedsAll', 'shBedsAll', 'rrhBedsAll', 'pshBedsAll', 'ophBedsAll',
                      'tempBedsNoChildren', 'permBedsNoChildren','esBedsNoChildren', 'thBedsNoChildren', 'shBedsNoChildren', 'rrhBedsNoChildren', 'pshBedsNoChildren', 'ophBedsNoChildren',
                      'tempBedsWithChildren', 'permBedsWithChildren','esBedsWithChildren', 'thBedsWithChildren', 'shBedsWithChildren', 'rrhBedsWithChildren', 'pshBedsWithChildren', 'ophBedsWithChildren',
                      'tempBedsOnlyChildren', 'permBedsOnlyChildren','esBedsOnlyChildren', 'thBedsOnlyChildren', 'shBedsOnlyChildren', 'rrhBedsOnlyChildren', 'pshBedsOnlyChildren', 'ophBedsOnlyChildren',
                      'tempUnitsWithChildren', 'permUnitsWithChildren','esUnitsWithChildren', 'thUnitsWithChildren', 'shUnitsWithChildren', 'rrhUnitsWithChildren', 'pshUnitsWithChildren', 'ophUnitsWithChildren',
                      'esBedsNonDv', 'thBedsNonDv', 'shBedsNonDv', 'rrhBedsNonDv', 'pshBedsNonDv', 'ophBedsNonDv',
                      'esBedsOverflow', 'esBedsSeasonal',
                      'esBedsDedVet', 'thBedsDedVet', 'shBedsDedVet', 'rrhBedsDedVet', 'pshBedsDedVet', 'ophBedsDedVet',
                      'esBedsDedYouth', 'thBedsDedYouth', 'shBedsDedYouth', 'rrhBedsDedYouth', 'pshBedsDedYouth', 'ophBedsDedYouth',
                      'pshBedsDedCh']]
```
