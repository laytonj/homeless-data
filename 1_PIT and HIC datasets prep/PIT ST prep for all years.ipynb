## Format, Clean-Up

*An extra row was created full of NaN. Remove this last empty row.*

```python
for file in pitFiles:
    if pd.isnull(file.index[-1]):
        file.drop(file.index[-1], inplace=True)
pit2017st.tail(n=3)
```

*Remove the Total row*

```python
for file in pitFiles:
    file.drop('Total', inplace=True)
```

*Remove the , YYYY from each column header*

```python
i = 0
for file in pitFiles:
    year = str(years[i])
    replaceThis = ', '+year
    file.columns = file.columns.map(lambda x: x.replace(replaceThis,''))
    i+=1
pit2017st.tail(n=3)
```

*Add a column for year*

```python
i = 0
for file in pitFiles:
    year = years[i]
    file.insert(0,'Year',year)
    i+=1
pit2017st.head(n=3)
```

*Format interior data as integers.*

# Create an all-years PIT DataFrame

*Combine all years*

```python
pitAllSt = pd.concat(pitFiles, axis=0, join="outer", levels=("Year"))
pitAllSt.head(n=3)
#note it sorted columns in alphabetical order
#note that columns including NaN cannot be type 'int' and become 'float'. See http://pandas.pydata.org/pandas-docs/stable/gotchas.html#na-type-promotions
```

*For earlier years when all chronically homeless were individuals, fill in total chronically homeless as individuals*

```python
pitAllSt.loc[['CA'],['year','totalChronic','shelChronic','unshelChronic','totalChronicIndiv','shelChronicIndiv','unshelChronicIndiv']]
```

```python
pitAllSt['totalChronic'] = np.where(pitAllSt['totalChronic'].isnull(), pitAllSt['totalChronicIndiv'], pitAllSt['totalChronic'])
pitAllSt['shelChronic'] = np.where(pitAllSt['shelChronic'].isnull(), pitAllSt['shelChronicIndiv'], pitAllSt['shelChronic'])
pitAllSt['unshelChronic'] = np.where(pitAllSt['unshelChronic'].isnull(), pitAllSt['unshelChronicIndiv'], pitAllSt['unshelChronic'])

pitAllSt.loc[['CA'],['year','totalChronic','shelChronic','unshelChronic','totalChronicIndiv','shelChronicIndiv','unshelChronicIndiv']]
```

*Create non-CH and non-Vets columns*

```python
pitAllSt['totalNonChronic'] = pitAllSt['totalH']-pitAllSt['totalChronic']
pitAllSt['shelNonChronic'] = pitAllSt['shelH']-pitAllSt['shelChronic']
pitAllSt['unshelNonChronic'] = pitAllSt['unshelH']-pitAllSt['unshelChronic']

pitAllSt.tail(n=3)
```

```python
pitAllSt['totalNonVet'] = pitAllSt['totalH']-pitAllSt['totalVet']
pitAllSt['shelNonVet'] = pitAllSt['shelH']-pitAllSt['shelVet']
pitAllSt['unshelNonVet'] = pitAllSt['unshelVet']-pitAllSt['unshelVet']

pitAllSt.tail(n=3)
```
