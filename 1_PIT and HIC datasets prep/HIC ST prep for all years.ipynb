*Remove commas, convert to numeric*

```python
def removeComma(x):
    if type(x) == str:
        x = x.replace(',', '')
        if x.isnumeric() == False:
            x = np.nan
    return float(x)

'''As seen with the PIT prep syntax, using looping through the files to implement applymap doesn't seem to work. 
Not sure why.'''
```

# Create an all-years HIC DataFrame

*Combine all years*

```python
hicAllSt = pd.concat(hicFilesSt, axis=0, join="outer")
hicAllSt.head(3)
#note it sorted columns in alphabetical order
```

```python
hicAllSt.columns.values
```

*Create columns grouping temporary beds and permanent beds*

```python
hicAllSt['permBedsAll'] = hicAllSt.fillna(0)['pshBedsAll'] + hicAllSt.fillna(0)['rrhBedsAll']
hicAllSt['permBedsWithChildren'] = hicAllSt.fillna(0)['pshBedsWithChildren'] + hicAllSt.fillna(0)['rrhBedsWithChildren']
hicAllSt['permBedsNoChildren'] = hicAllSt.fillna(0)['pshBedsNoChildren'] + hicAllSt.fillna(0)['rrhBedsNoChildren']
hicAllSt['permBedsOnlyChildren'] = hicAllSt.fillna(0)['pshBedsOnlyChildren'] + hicAllSt.fillna(0)['rrhBedsOnlyChildren']
hicAllSt['permUnitsWithChildren'] = hicAllSt.fillna(0)['pshUnitsWithChildren'] + hicAllSt.fillna(0)['rrhUnitsWithChildren']

hicAllSt['tempBedsAll'] = hicAllSt.fillna(0)['esBedsAll'] + hicAllSt.fillna(0)['thBedsAll'] + hicAllSt.fillna(0)['shBedsAll']
hicAllSt['tempBedsWithChildren'] = hicAllSt.fillna(0)['esBedsWithChildren'] + hicAllSt.fillna(0)['thBedsWithChildren'] + hicAllSt.fillna(0)['shBedsWithChildren']
hicAllSt['tempBedsNoChildren'] = hicAllSt.fillna(0)['esBedsNoChildren'] + hicAllSt.fillna(0)['thBedsNoChildren'] + hicAllSt.fillna(0)['shBedsNoChildren']
hicAllSt['tempBedsOnlyChildren'] = hicAllSt.fillna(0)['esBedsOnlyChildren'] + hicAllSt.fillna(0)['thBedsOnlyChildren'] + hicAllSt.fillna(0)['shBedsOnlyChildren']
hicAllSt['tempUnitsWithChildren'] = hicAllSt.fillna(0)['esUnitsWithChildren'] + hicAllSt.fillna(0)['thUnitsWithChildren'] + hicAllSt.fillna(0)['shUnitsWithChildren']
```

*Create a column for sum of all beds*

```python
hicAllSt['allBeds'] = hicAllSt['permBedsAll'] + hicAllSt['tempBedsAll']
```

*Revise the order of columns because they had been automatically sorted alphabetically above*

```python
hicAllSt = hicAllSt[['year', 'allBeds',
                      'tempBedsAll', 'permBedsAll', 'esBedsAll', 'thBedsAll', 'shBedsAll', 'rrhBedsAll', 'pshBedsAll', 'ophBedsAll',
                      'tempBedsNoChildren', 'permBedsNoChildren','esBedsNoChildren', 'thBedsNoChildren', 'shBedsNoChildren', 'rrhBedsNoChildren', 'pshBedsNoChildren', 'ophBedsNoChildren',
                      'tempBedsWithChildren', 'permBedsWithChildren','esBedsWithChildren', 'thBedsWithChildren', 'shBedsWithChildren', 'rrhBedsWithChildren', 'pshBedsWithChildren', 'ophBedsWithChildren',
                      'tempBedsOnlyChildren', 'permBedsOnlyChildren','esBedsOnlyChildren', 'thBedsOnlyChildren', 'shBedsOnlyChildren', 'rrhBedsOnlyChildren', 'pshBedsOnlyChildren', 'ophBedsOnlyChildren',
                      'tempUnitsWithChildren', 'permUnitsWithChildren','esUnitsWithChildren', 'thUnitsWithChildren', 'shUnitsWithChildren', 'rrhUnitsWithChildren', 'pshUnitsWithChildren', 'ophUnitsWithChildren',
                      'esBedsNonDv', 'thBedsNonDv', 'shBedsNonDv', 'rrhBedsNonDv', 'pshBedsNonDv', 'ophBedsNonDv',
                      'esBedsOverflow', 'esBedsSeasonal',
                      'esBedsDedVet', 'thBedsDedVet', 'shBedsDedVet', 'rrhBedsDedVet', 'pshBedsDedVet', 'ophBedsDedVet',
                      'esBedsDedYouth', 'thBedsDedYouth', 'shBedsDedYouth', 'rrhBedsDedYouth', 'pshBedsDedYouth', 'ophBedsDedYouth',
                      'pshBedsDedCh']]
```
