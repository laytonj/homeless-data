## Add column noting caution for comparison
*If more than 20 percent increase or decrease in CoC population as a result of the geography change, exclude*

```python
compareOnlyUntil = {'AR-503':2013, 'AR-508':2014, 'AR-512':2010, 'CA-515':2009, 'CA-516':2015, 
                    'CA-523':2015, 'CA-527':2015, 'CA-529':2015, 'CA-615':2015, 'GA-500':2013,
                    'GA-502':2013, 'GA-506':2013, 'IL-506':2013, 'IL-512':2013, 'LA-509':2015,
                    'MA-504':2012, 'MA-507':2012, 'MA-517':2009, 'MA-518':2009, 'NY-501':2011,
                    'NY-505':2009, 'NY-511':2013, 'NY-518':2012, 'NY-520':2012, 'NY-522':2008,
                    'PA-512':2011, 'TX-503':2008
                   }
```

```python
#Make temporary values being combination of coc number and year
pitHicAllCocClean.reset_index(inplace=True)
pitHicAllCocClean['comparable'] = pitHicAllCocClean['year'].map(str) + pitHicAllCocClean['cocNum']

pitHicAllCocCleanGroups.reset_index(inplace=True)
pitHicAllCocCleanGroups['comparable'] = pitHicAllCocCleanGroups['year'].map(str) + pitHicAllCocCleanGroups['cocNum']

pitHicAllCocClean.head(3)
```

```python
#define function that checks for coc number in dictionary, compares year to year in dictionary
def checkIfCompareCaution(x):
    coc = str(x[4:])
    year = int(x[:4])
    if coc in compareOnlyUntil:
        if compareOnlyUntil[coc] > year:
            return False
        else:
            return True
    else:
        return True
```

```python
#apply the function
pitHicAllCocClean['comparable'] = pitHicAllCocClean['comparable'].map(lambda x: checkIfCompareCaution(x))
pitHicAllCocClean['cocComparableTo'] = pitHicAllCocClean['cocNum'].apply(lambda x: compareOnlyUntil[x] if x in compareOnlyUntil else np.nan)
pitHicAllCocClean.set_index('cocNum', inplace=True)

pitHicAllCocCleanGroups['comparable'] = pitHicAllCocCleanGroups['comparable'].map(lambda x: checkIfCompareCaution(x))
pitHicAllCocCleanGroups['cocComparableTo'] = pitHicAllCocCleanGroups['cocNum'].apply(lambda x: compareOnlyUntil[x] if x in compareOnlyUntil else np.nan)
pitHicAllCocCleanGroups.set_index('cocNum', inplace=True)
```
