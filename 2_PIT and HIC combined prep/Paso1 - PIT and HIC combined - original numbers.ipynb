# Merge ST

```python
pitAllSt.reset_index(inplace=True)
hicAllSt.reset_index(inplace=True)
```

```python
pitAllSt.set_index(['st','year'], inplace=True)
hicAllSt.set_index(['st','year'], inplace=True)
pitAllSt.head(3)
```

```python
pitHicAllSt = pitAllSt.merge(hicAllSt, how='outer', left_index=True, right_index=True)
pitHicAllSt.tail(3)
```

# Merge CoC

*Reset index, standardize 'st' name, and set multi-index for the merger*

```python
pitAllCoc.reset_index(inplace=True)
hicAllCoc.reset_index(inplace=True)
```

```python
pitAllCoc = pitAllCoc.rename(columns = {"ST":"st"})
hicAllCoc = hicAllCoc.rename(columns = {"ST":"st"})
hicAllCoc.head(3)
```

```python
pitAllCoc.set_index(['cocNum','year'], inplace=True)
hicAllCoc.set_index(['cocNum','year'], inplace=True)
hicAllCoc.head(3)
```

*MERGE!*

```python
pitHicAllCoc = pitAllCoc.merge(hicAllCoc, how='outer', left_index=True, right_index=True)
```

```python
pitHicAllCoc.head(3)
```

*Result was st_x and st_y because some CoCs that appear on the HIC in a given year don't appear on the PIT*<br>
*Create a column that flags if they don't match*

```python
pitHicAllCoc['cocOnBoth'] = np.where(pitHicAllCoc['st_x'] == pitHicAllCoc['st_y'], 'True', 'False')
pitHicAllCoc.head(3)
```

*Reset the index to be only cocNum*

```python
pitHicAllCoc.reset_index(inplace=True)
pitHicAllCoc.set_index('cocNum', inplace=True)
```
